```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
```

```python
data = pd.read_csv("data/percents_ops.txt")
dataframe = pd.DataFrame(data)
pivot_1, pivot_2, nb_ops = dataframe["pivot_1"], dataframe["pivot_2"], dataframe["nb_ops"]

plt.figure()
plt.plot(pivot_2, nb_ops, label= fr"$pivot_1$={pivot_1[0]}", color= 'orange')
plt.legend()
plt.ylabel("Opérations")
plt.axhline(5500, label= "Limite", ls= '--', color= 'r')
plt.xlabel(r"$Pivot_2$")
plt.title(rf"Min : {np.min(nb_ops)} operations, at $pivot_2$={pivot_2[np.argmin(nb_ops)]}%")
plt.scatter(pivot_2[np.argmin(nb_ops)], np.min(nb_ops), label= "Min", color= 'r', marker= 'x')
plt.grid()
plt.show()

plt.figure()
plt.plot(pivot_2 - pivot_1, nb_ops)
plt.show()
```
